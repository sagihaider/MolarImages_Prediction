```python
DATA_DIR = ('LowerMolar_Cleaned')
# Check whether the specified path exists or not
isExist = os.path.exists(DATA_DIR)
if not isExist:
  # Create a new directory because it does not exist 
  os.makedirs(DATA_DIR)
  print("The new directory is created!")
```

```python
base_image_dir = '/content/LowerMolar_Raw'
```

```python
image_list = os.listdir(base_image_dir)
len(image_list)
```

```python
data_df = pd.DataFrame(columns=['patient_num', 'image_num', 'img_path', 'age'])
```

```python
for img in tqdm(image_list):
    path_split = img.split('_')
    print(path_split)
    data_df = data_df.append({
        'patient_num' :int(path_split[1]),
        'image_num' :int(path_split[2][-1]),
        'img_path' : os.path.join(base_image_dir, img),
        'age' :int(path_split[3].replace('age','').replace('.jpg',''))
    }, ignore_index=True)
```

```python
data_df
```

```python
data_df.to_csv(f'{DATA_DIR}/all_data.csv', index=False)
```

```python
all_patient_nums = data_df['patient_num'].unique()
```

```python
train_idx, val_idx, test_idx = np.split(np.random.permutation(all_patient_nums), [int(.8 * len(all_patient_nums)), int(.9 * len(all_patient_nums))])
```

```python
train_df = data_df[data_df['patient_num'].isin(train_idx)]
val_df = data_df[data_df['patient_num'].isin(val_idx)]
test_df = data_df[data_df['patient_num'].isin(test_idx)]
```

```python
train_df.to_csv(f'{DATA_DIR}/train.csv', index=False)
val_df.to_csv(f'{DATA_DIR}/val.csv', index=False)
test_df.to_csv(f'{DATA_DIR}/test.csv', index=False)
```
